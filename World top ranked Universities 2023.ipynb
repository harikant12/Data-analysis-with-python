# Australian universities

```python
data_a = data[(data['location'] == 'Australia')]
data_a
```

```python
data.loc[data.location.isin(["Australia"])]
```

```python
data_a.groupby(['Rank', 'institution', 'location'])['ger rank', 'score scaled'].describe()
```

```python
data_a.info()
```

```python
plt.figure(figsize=(40, 20))
x = data_a['institution']
y = data_a['Rank']
plt.hist(x, bins=30, color='red', edgecolor='black',
         linestyle='--', alpha=0.5)
 
# Adding title to the plot
plt.title("Uni Ranking")
 
# Adding label on the y-axis
plt.ylabel('Ranking')
 
# Adding label on the x-axis
plt.xlabel('Universities')
 
plt.show()
```

```python
data_a.head()
```

```python
data_a.describe()
```

```python
plt.figure(figsize=(30, 20))
x = data_a['Rank']
y = np.vstack([data_a['institution'], 
                   data_a['ger rank'], 
                      data_a['score scaled']])

labels = ['institution', 'ger rank', 'scaled score']

colors = ['sandybrown', 'tomato', 'skyblue']

plt.stackplot(x,y, labels = labels,
             colors = colors)
 
plt.legend(loc=2)
 
plt.show()
```

```python
plt.figure(figsize=(25, 17))
sns.set()
sns.barplot(data_a['Rank'], data_a['institution'])
```
